```python
#Задание 1
"""
Скачайте с сайта grouplens.org...movielens/ датасет любого размера. 
Определите какому фильму было выставлено больше всего оценок 5.0.
"""
import pandas as pd
ratings = pd.read_csv('C:\\Users\\Оляша\\Desktop\\Netology\\DZ\\2_Numpy_pandas\\2_Pandas_base\\ml-latest-small\\ratings.csv')
movies = pd.read_csv('C:\\Users\\Оляша\\Desktop\\Netology\\DZ\\2_Numpy_pandas\\2_Pandas_base\\ml-latest-small\\movies.csv')
# создаем объединенный dataframe, чтобы связать (по movieId) названия фильмов с рейтинговыми данными
joined = ratings.merge(movies, on='movieId', how='left')
# определяем название фильма, у которого больше всего 5
joined.title[joined.rating == 5.0].value_counts().head(1) 
```

```python
#Задание 2
"""
По данным файла power.csv посчитайте суммарное потребление 
стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. 
Не учитывайте в расчетах отрицательные значения quantity.
"""
data = pd.read_csv('C:\\Users\\Оляша\\Desktop\\Netology\DZ\\2_Numpy_pandas\\2_Pandas_base\\power.csv')
#создаем отфильтрованный по заданным условиям dataframe
filtered_data = (data[data['country'].isin(['Latvia', 'Lithuania', 'Estonia'])& 
                      data['year'].isin(range(2005, 2011))&                     
                      data['category'].isin([12, 4, 21])])
#группируем по странам и категориям с вычислением суммы потребления (для значений выше 0)
(filtered_data[ filtered_data.quantity > 0 ].
 groupby(['country', 'category']).
 sum().pop('quantity'))
```

```python
#Задание 3
"""
Выберите страницу любого сайта с табличными данными. Импортируйте таблицы в pandas dataframe.
Примеры страниц (необязательно брать именно эти):
https://fortrader.org/quotes
www.finanz.ru...om-vremeni
"""
import re
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

url = 'https://fortrader.org/quotes'
driver = webdriver.Firefox(executable_path=r'C:\Program Files (x86)\Mozilla Maintenance Service\geckodriver.exe')
driver.get(url)
#находим все таблицы на сайте
tables = driver.find_elements_by_tag_name('table')
i = 0
#перебираем в списке все таблицы для импорта в pandas dataframe
for i in range(1, len(tables)):
#считываем таблицу со всеми атрибутами ('outerHTML')
    html = tables[i].get_attribute('outerHTML')
    #присваиваем в качестве индекса значение первого столбца для красивого вывода
    df  = pd.read_html(html, index_col = 0) 
    #запрещаем переносы для красивого вывода
    pd.options.display.expand_frame_repr = False
    print(df)
driver.close()
```
