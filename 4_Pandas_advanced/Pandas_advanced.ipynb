```python
# Задание 1

""" Для датафрейма log из материалов занятия создайте столбец source_type по 
следующим правилам:

    если источник traffic_source равен yandex или google, то в source_type ставится organic
    для источников paid и email из России - ставим ad
    для источников paid и email не из России - ставим other
    все остальные варианты берем из traffic_source без изменений
   """


import pandas as pd
#import time #для замера времени исполнения скрипта
log = pd.read_csv('C:\\Users\\Оляша\\Desktop\\Netology\\DZ\\2_Numpy_pandas\\4_Pandas_advanced\\visit_log.csv', sep=';')

start_time1 = time.time()#для замера времени исполнения скрипта

#условия исходя из задания
cond1 = (log.traffic_source == 'yandex') | (log.traffic_source == 'google')
cond2 = ((log.traffic_source == 'paid') | (log.traffic_source == 'email'))
cond3 = (log.region == 'Russia')

#заполнение нового столбца source_type исходя из условий
log.loc[cond1,'source_type'] = 'organic'
log.loc[cond2 & cond3, 'source_type'] = 'ad' 
log.loc[cond2 & (~cond3), 'source_type'] ='other'

#замена оставшихся NaN значений на данные из traffic_source
log.source_type.fillna(log.traffic_source, inplace = True)

#print(time.time()- start_time1) #для замера времени исполнения скрипта

log.head(10)
```

```python
# Задание 1 ('медленное' решение)

import pandas as pd
log = pd.read_csv('C:\\Users\\Оляша\\Desktop\\Netology\\DZ\\2_Numpy_pandas\\4_Pandas_advanced\\visit_log.csv', sep=';')

#start_time = time.time() #для замера времени исполнения скрипта



organic_sources, ad_sources  = set(['yandex', 'google']), set(['paid', 'email'])

def SourceType(row):
    '''Эта функция классифицирует типы трафика исходя из источников трафика'''
    
    if row.traffic_source in organic_sources:
        return 'organic'
    elif row.traffic_source in ad_sources and row.region == 'Russia':
        return 'ad'
    elif row.traffic_source in ad_sources and row.region != 'Russia':
        return 'other'
    
    return row.traffic_source

    
log['source_type'] = log.apply(SourceType, axis = 1)

#print(time.time()-start_time) #для замера времени исполнения скрипта
log.head(10)
```

```python
#Задание 3

"""Используйте файл с оценками фильмов ml-latest-small/ratings.csv. 
Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. 
Под временем жизни понимается разница между 
максимальным и минимальным значением столбца timestamp для данного значения userId."""

ratings = pd.read_csv('C:\\Users\\Оляша\\Desktop\\Netology\\DZ\\2_Numpy_pandas\\4_Pandas_advanced\\ratings.csv')

#группировка по количеству пользователей проставленных рейтингов с укаазанием макс и мин timestamp
ratings_grouped = ratings.groupby(['userId']).agg({'rating': 'count','timestamp':['max','min']}) 

#подсчет времени жизни для каждого пользователя
ratings_grouped['diff'] =  ratings_grouped['timestamp']['max'] - ratings_grouped['timestamp']['min']

#подсчет среднего времени жизни пользователей, которые выставили более 100 оценок. 
ratings_grouped[ratings_grouped['rating']['count'] > 100]['diff'].mean() / 24 / 3600

```

```python
#Задание 4

"""    Дана статистика услуг перевозок клиентов компании по типам 
(см. файл “Python_13_join.ipynb” в разделе 
Материалы для лекции «Продвинутый pandas» ---- Ноутбуки к лекции «Продвинутый pandas»).
    
    Необходимо сформировать две таблицы:

    таблицу с тремя типами выручки для каждого client_id без указания адреса клиента
    аналогичную таблицу по типам выручки с указанием адреса клиента

Обратите внимание, что в процессе объединения таблиц данные не должны теряться."""

from functools import reduce

rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)


auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)


air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)

client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)



dfs = [client_base, rzd, auto, air]

#кумулятивное объединение датафреймов по client_id методом outer join
dfs_merged = reduce(lambda  left, right: pd.merge(left, right,on = ['client_id'], how = 'outer'), dfs).fillna(0)

print("Данные по выручке для каждого клиента (без указания адреса)\n\n",dfs_merged.drop(columns = 'address'))
print("\nДанные по выручке для каждого клиента (c указанием адреса)\n\n", dfs_merged)
```
