```python
#Задание 1**
"""
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class"""


import pandas as pd
ratings = pd.read_csv('C:\\Users\\Оляша\\Desktop\\Netology\\DZ\\2_Numpy_pandas\\3_Pandas_functions\\ratings.csv')
movies = pd.read_csv('C:\\Users\\Оляша\\Desktop\\Netology\\DZ\\2_Numpy_pandas\\3_Pandas_functions\\movies.csv')
# создаем объединенный dataframe, чтобы связать (по movieId) названия фильмов с рейтингом
joined = ratings.merge(movies, on='movieId', how='left')

def rating(row):
    """
    Эта функция определяет к какому классу относится фильм на основе его рейтинга
    
    """
    
    if (row <= 2):
        return 'низкий рейтинг' 
    elif(row <= 4):
        return 'средний рейтинг'
    
    return 'высокий рейтинг'

joined['class'] = joined['rating'].apply(rating)
joined.head()
```

```python
# Задание 2
"""Используем файл keywords.csv.
Необходимо написать гео-классификатор, который каждой строке сможет выставить 
географическую принадлежность определенному региону. 
Т. е. если поисковый запрос содержит название города региона, 
то в столбце ‘region’ пишется название этого региона. 
Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

geo_data = {

'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']

}
Результат классификации запишите в отдельный столбец region."""

keywords = pd.read_csv('C:\\Users\\Оляша\\Desktop\\Netology\\DZ\\2_Numpy_pandas\\3_Pandas_functions\\keywords.csv')

geo_data = {

'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']

}

def geoClass(row):
    
    """Эта функция возвращает регион по названию города в составе запроса 
    (ключ по значению словаря geo_data )
    """
    
    for region, cities in geo_data.items():
        for city in cities:
            if city in row: #поиск названия города в запросе
                 return region
    
    return 'undefined'

keywords['region'] = keywords['keyword'].apply(geoClass)

keywords[keywords['region'] != 'undefined'].head() #вывод только где есть регион
```
